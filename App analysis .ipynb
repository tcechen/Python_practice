```python
#input two data 'Applestore.csv' and 'googleplaystore.csv' and turn the data into list, seperate the header
from csv import reader
open_file=open('AppleStore.csv',encoding='utf8')
read_file=reader(open_file)
ios=list(read_file)
ios_header=ios[0]
ios=ios[1:]

open_file=open('googleplaystore.csv',encoding='utf8')
read_file=reader(open_file)
android=list(read_file)
android_header=android[0]
android=android[1:]
```

```python
#find duplicate data
duplicate_apps=[]
unique_apps=[]
for app in android:
    name=app[0]
    if name in unique_apps:
        duplicate_apps.append(name)
    else:
        unique_apps.append(name)
        
print("Number of duplicate apps:",len(duplicate_apps))
print('\n')
print('Example of duplicate apps:',duplicate_apps[:15])
```

```python
#use review_max to keep the highest number of review
print("Expected length:",len(android)-1181)
review_max={}
for row in android:
    name=row[0]
    n_reviews=float(row[3])
    if name in review_max and review_max[name]<n_reviews:
        review_max[name]=n_reviews
    elif name not in review_max:
        review_max[name]=n_reviews
print("Actual length:",len(review_max))
```

```python
#print out the data
def explore_data(dataset, start, end, rows_and_columns=False):
    dataset_slice=dataset[start:end]
    for row in dataset_slice:
        print(row)
        print('\n')
    if rows_and_columns:
        print('Number of rows:',len(dataset))
        print('Number of columns:',len(dataset[0]))

        
print(ios_header)
print('\n')
explore_data(ios,0,5,True)


print(android_header)
print('\n')
explore_data(android,0,5,True)
```

```python
#delete duplicate row in android
android_clean=[]
already_added=[]
for row in android:
    name=row[0]
    n_review=float(row[3])
    if (review_max[name]==n_review) and (name not in already_added):
        android_clean.append(row)
        already_added.append(name)
explore_data(android_clean,0,3,True) 
```

```python
#Remove non-English app
#accorgin to ASCII: English characters 0-127
print(ord('a'))
print(ord('Êàë'))
def is_English(string):
    count=0
    for character in string: #count if there's more than three non_english characters
        if ord(character)>127:
            count+=1
        
    if count>3:
        return False  
    else:
        return True

print(is_English('Instagram'))
print(is_English('Áà±Â•áËâ∫PPS -„ÄäÊ¨¢‰πêÈ¢Ç2„ÄãÁîµËßÜÂâßÁÉ≠Êí≠'))
print(is_English('Docs To Go‚Ñ¢ Free Office Suite'))
print(is_English('Instachat üòú'))
```

```python
#consider only English app
android_English=[]
ios_English=[]
for app in android_clean:
    name=app[0]   #only consider the name of each data
    if is_English(name):
        android_English.append(app)

for app in ios:
    name=app[2]
    if is_English(name):
        ios_English.append(app)

explore_data(android_English,0,3,True)
print('\n')
explore_data(ios_English,0,3,True)
```

```python
#consider only free app
android_final=[]
ios_final=[]
for app in android_English:
    price=app[7]
    if price =='0':
        android_final.append(app)
for app in ios_English:
    price=app[5]
    if price == '0':
        ios_final.append(app)
print(len(android_final))
print(len(ios_final))
```

```python
#find the frequency of one specific row 
def freq_genre(dataset, index):
    table = {}
    total = 0
    
    for row in dataset:
        total += 1
        value = row[index]
        if value in table:
            table[value] += 1
        else:
            table[value] = 1
    
    table_freq = {}
    for key in table:
        percentage = (table[key] / total) * 100
        table_freq[key] = percentage 
    return table_freq

def display_freq(dataset, index):
    table=freq_genre(dataset, index) #call freq_genre function
    table_display=[]
    for key in table:
        key_val_tuple=(table[key],key)
        table_display.append(key_val_tuple)
        
    #change dictionary to tuple(consider the value)    
    table_sorted=sorted(table_display, reverse=True)  
    
    for entry in table_sorted:
        print(entry[1],':',entry[0])

display_freq(ios_final, -5)
print('\n')
display_freq(android_final,1)
print('\n')
display_freq(android_final,-4)
```

```python
#calculate average number of user's ratings per app genre of ios
genre_ios=freq_genre(ios_final,-5)
#genre_ios is the percentage of genre in ios

for genre in genre_ios:
    total=0
    len_genre=0
    for app in ios_final:
        genre_app=app[-5]  #app[-5] is the genre
        if genre_app == genre:
            n_ratings=float(app[6]) #app[6] is the rating of app
            total+=n_ratings
            len_genre+=1
    avg_ratings=total/len_genre
    print(genre, " : ", avg_ratings)
#find out Navigation is the largest number of user review


print('\nThe app in Navigation:\n')
for app in ios_final:
    if app[-5]=='Navigation':
        print(app[2],':',app[6])

print('\nThe app in Reference:\n')
for app in ios_final:
    if app[-5]=='Reference':
        print(app[2],':',app[6])
```

```python
display_freq(android_final,5) #the install of column

category_android=freq_genre(android_final,1)
print('\n\n')
for category in category_android:
    total=0
    len_category=0
    for app in android_final:
        category_app=app[1]
        if category_app==category:
            n_install=app[5]
            n_install=n_install.replace(',','')
            n_install=n_install.replace('+','')
            total+=float(n_install)
            len_category+=1
    avg_install=total/len_category
    print(category,':',avg_install)
```
